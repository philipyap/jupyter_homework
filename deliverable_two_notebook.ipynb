```python
num = [3, 78, 23, 466, 324, 2, 12, 546, 17, 546, 17, 56, 23, 123, 343, 87, 94, 45, 6, 9, 234, 36]
```

```python
def merge_sort(ls):
    if len(ls) <= 1:
        return ls
    mid = len(ls)
    left = ls[:mid]
    print('left list:', left)
    right = ls[mid:]
    print('right list:', right)
    
    left_sorted = merge_sort(left)
    right_sorted = merge_sort(right)
    
    return merge(left_sorted, right_sorted)
```

```python
def merge(ls1, ls2):
    print('ls1: ', ls1)
    print('ls2: ', ls2)
    sort = []
    while len(ls1) and len(ls2):
        if ls1[0] > ls2[0]:
            sort.append(ls2.pop(0))
        else:
            sort.append(ls1.pop(0))
    sort = sort + ls1 + ls2
    print('sorted list: ', sort)
    return(sort)
merge_sort(num)
```
